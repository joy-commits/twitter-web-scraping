```python
for tweet in tweets:
    print(tweet.text)
```

```python
#saving in the CSV format

import csv

with open('tweets.csv', 'w', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['id', 'created_at', 'text'])
    
    for tweet in tweets:
        writer.writerow([tweet.id, tweet.created_at, tweet.text])
```
